# Loading files
We have two folders, one with AACT (tabulated ClinicalTrials.gov data) and one with HINT/TOP data. We can also load TDC data.

## From AACT
Here I loaded the countries, studies (date) and participant tables. We can load other tables for our dataset

```python
countries = pd.read_csv(deep_learning_dir + '/data/AACT_202203/countries.txt', sep='|')
```

```python
countries
```

```python
studies = pd.read_csv(deep_learning_dir + '/data/AACT_202203/studies.txt', sep ='|')
studies
```

```python
studies = studies[['nct_id', 'last_update_submitted_date']]
```

```python
participants = pd.read_csv(deep_learning_dir + '/data/AACT_202203/outcome_counts.txt', sep='|')
participants = participants[participants['units']=='Participants']
participants = participants[['nct_id', 'count']]
```

```python
participants
```

## From HINT
I only loaded phase 3 test dataset but we can load all of them and do the same processing

```python
trial_phase_3_test = pd.read_csv(deep_learning_dir + '/data/clinical_trial_outcome/data/phase_III_test.csv')
```

## Merging files

```python
trial_phase_3_test = trial_phase_3_test.rename(columns={"nctid": "nct_id"})
```

```python
trial_phase_3_test
```

```python
trial_phase_3_test = trial_phase_3_test.merge(studies, how='left')
trial_phase_3_test = trial_phase_3_test.merge(countries, how='left')
trial_phase_3_test = trial_phase_3_test.merge(participants, how='left')
trial_phase_3_test = trial_phase_3_test[['nct_id','status','phase','diseases','icdcodes','drugs','smiless','last_update_submitted_date','name','count','label']]
trial_phase_3_test = trial_phase_3_test.rename(columns={'last_update_submitted_date':'completion_date', 'name':'country', 'count':'participant_count', 'label':'outcome'})
trial_phase_3_test[~trial_phase_3_test['participant_count'].isna()]
```
